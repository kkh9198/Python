## 페이지별 데이터 수집 함수 만들기

```python
# 종목 번호를 이용해 page에 따라 데이터를 읽어오는 함수
# """ 는 이 두개 사이의 행들은 주석 처리되며, 함수의 docstring 으로 사용됩니다.

def get_day_board_list(item_code, page_no):
    """
    일자별 종목 게시판 자료를 수집
    """ 
    url = f"https://finance.naver.com/item/board.nhn?code={item_code}&page={page_no}"
    
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.150 Safari/537.36'}
    response = requests.get(url, headers=headers)
    
    html = bs(response.text, "lxml")
    table = html.select("table")
    table = pd.read_html(str(table))
    
    table[1]=table[1].drop('Unnamed: 6',axis=1)
    temp = table[1].dropna()
    return temp
```

```python
# 함수가 잘 만들어졌는지 확인하기
get_day_board_list(item_code, 1)
```

## 반복문을 통한 전체 일자 데이터 수집하기
* (주의) 기간이 긴 데이터를 수집할때는 서버에 부담을 주지 않기 위해 time.sleep()값을 주세요.

```python
import time
# web page 시작번호
page_no = 1
# 데이터를 저장할 빈 변수 선언
item_board_list = []

while True:
    print(page_no)
    temp = get_day_board_list(item_code, page_no)
    item_board_list.append(temp)
    
    page_no = page_no + 1
    time.sleep(0.1)
    
   
    if page_no > 10:
        break;
```
